```python
%%bash
set -e

copy_tree() {
  local src_root="$1"
  local dst_root="$2"

  find "$src_root" -type f | while read -r file; do
    rel="${file#$src_root/}"
    target="$dst_root/$rel"

    echo "Copying $file -> $target"
    mkdir -p "$(dirname "$target")"
    cp "$file" "$target"
  done
}

PATCH_ROOT="/workspaces/OpenICU.debug-mimic-iv-meds-etl/patches/origins_venv"

MIMIC_DIR=$(/workspaces/.venv/bin/python -c "import os, MIMIC_IV_MEDS; print(os.path.dirname(MIMIC_IV_MEDS.__file__))")
MEDS_DIR=$(/workspaces/.venv/bin/python -c "import os, MEDS_transforms; print(os.path.dirname(MEDS_transforms.__file__))")

copy_tree "$PATCH_ROOT/MIMIC_IV_MEDS" "$MIMIC_DIR"
copy_tree "$PATCH_ROOT/MEDS_transforms" "$MEDS_DIR"
```
